# **Осцилације нивоа течности у водостану**

```python
import numpy as np
import matplotlib.pyplot as plt
#np.set_printoptions(formatter={'float': '{: 0.5f}'.format})

pi = np.pi
g = 9.81
Q0 = 30.
Q1 = 15

Lt = 6000.
Dt = 3.
Dv = 8.
lam = 0.01

Av = Dv**2 * pi/4 ## povrsina vodostana
At = Dt**2 * pi/4 ## povrsina tunela

deltaT = 1

zv0 = -18.36
Qt0 = 30.

time = []
zv = []
Qt = []

zPrethodno = zv0
QtPrethodno = Q0

brojIteracija = 2000
for k in range(brojIteracija):
    zNovo = zPrethodno + deltaT/Av*(QtPrethodno - Q1)     
    QtNovo = QtPrethodno + deltaT*(g*At/Lt*(-0.5*(zNovo+zPrethodno)) - 2*lam/(Dt**3*pi)*QtPrethodno*QtPrethodno)
    time.append(k*deltaT)
    zv.append(zNovo)
    Qt.append(QtNovo)
    zPrethodno = zNovo
    QtPrethodno= QtNovo
    
    
## Nivo vode u vodostanu nakon uspostavljanja stacionarnog stanja
zStac = zv[brojIteracija-1]
print (zStac)
plt.figure(figsize=(16,8), dpi=100)

fig1 = plt.subplot(1,2,1)
fig1.set_title('Промена нивоа воде у водостану', fontsize=20)
fig1.grid(True)


fig2 = plt.subplot(1,2,2)
fig2.set_title('Промена протока у тунелу', fontsize=20)
fig2.grid(True)

fig1.plot(time, zv, label = "$z_v$", linewidth=2)
fig1.legend(fontsize = 16)


fig2.plot(time, Qt, label = "$Q_T$", linewidth=2)
fig2.legend(fontsize = 16)
```
