## **Пример прорачуна водоводне мреже $\boldsymbol{\varPi}$-методом**

Израчунати вредности пијезометарских кота у чворова за прстенасту мрежу дату на слици. Познати су вредности потрошњи у чворовима: $\dot{V}_{p2} = 0.5\,\mathrm{m^3/s}$, $\dot{V}_{p3} = 0.6\,\mathrm{m^3/s}$, $\dot{V}_{p5} = 0.4\,\mathrm{m^3/s}$, хидрауличке карактеристике цевовода дате у табели, као и вредност пијезометарске коте у чвору 0, $\varPi_0 = 50\,\mathrm{m}$. 

|$K_{01}$|$K_{03}$|$K_{12}$|$K_{14}$|$K_{25}$|$K_{34}$|$K_{45}$|
| --- | --- | --- | --- | --- | --- | --- |
| 22 | 12 | 11 | 15 | 16 | 9 | 30 |

**Решење.** У овој мрежи има укупно шест чворова, али је вредност пијезометарске коте у чвору 0 задата, тако да су непознате величине пијезометарске коте чворова 1-5. Дакле, имамо 5 непознатих величина, за које морамо да формирамо пет независних једначина. То ће бити систем линеаризованих једначина континуитета, изражених преко пијезометарских кота. Њен општи облик за произвољни $i$-ти чвор
\begin{equation}
  \sum_{j} \, \frac{\varPi_i^{(k+1)} -  \varPi_j^{(k+1)}}{K_{ij} |\dot{V}_{ij}^{(k)}|} = - 2\dot{V}_{ip} - \sum_j \, \dot{V}_{ij}^{(k)}
  \tag{1}
\end{equation}
Са индексом $ј$ су означени суседни чворови чвора $i$. Горњи индекси означавају итерације: тако индекс $(k)$ одговара текућим вредностима, док индекс $(k+1)$ означава вредности пијезометарких кота у наредној итерацији. Тако је пример, за чвор означен са **1:** $i=1$, док су $j=0, 2, 4$ и потрошња у чвору $\dot{V}_{p1} = 0\,\mathrm{m^3/s}$. Тако једначина (1) написана за чвор **1** је облика
\begin{equation}
 \frac{\varPi_1^{(k+1)} - \varPi_0}{K_{10}|\dot{V}_{10}^{(k)}|}
      + \frac{\varPi_1^{(k+1)} - 
        \varPi_2^{(k+1)}}{K_{12}|\dot{V}_{12}^{(k)}|} + \frac{\varPi_1^{(k+1)} -
        \varPi_4^{(k+1)}}{K_{14}|\dot{V}_{14}^{(k)}|} =
      -\left(\dot{V}_{10}^{(k)} + \dot{V}_{12}^{(k)} + \dot{V}_{14}^{(k)}\right)
\end{equation}
\begin{equation}
\left(\frac{1}{K_{10}|\dot{V}_{10}^{(k)}|} + \frac{1}{K_{12}|\dot{V}_{12}^{(k)}|} + \frac{1}{K_{14}|\dot{V}_{14}^{(k)}|} \right)\varPi_1^{(k+1)} - 
 \frac{1}{K_{12}|\dot{V}_{12}^{(k)}|}\,\,\varPi_{2}^{(k+1)} - \frac{1}{K_{14}|\dot{V}_{14}^{(k)}|}\,\, \varPi_4^{(k+1)} = \frac{\varPi_0}{K_{10}|\dot{V}_{10}^{(k)}|}  - \left(\dot{V}_{10}^{(k)} + \dot{V}_{12}^{(k)} + \dot{V}_{14}^{(k)}\right)
\end{equation}

\begin{equation}
a_{11}^{(k)} \, \varPi_1^{(k+1)} + a_{12}^{(k)} \,  \varPi_2^{(k+1)} + a_{14}^{(k)} \, \varPi_4^{(k+1)} = b_1^{(k)}\
\end{equation}

По истом принципу, написаћемо једначине континуитета и за остале чворове **2, 3, 4** и **5**, и на тај начин добити систем линеарних једначина - пет линеарних једначина са пет непознатих, које су могу представити на следећи начин
\begin{equation}
\left.
\begin{array}{r}
  a_{11}^{(k)} \, \varPi_1^{(k+1)} + a_{12}^{(k)} \,
  \varPi_2^{(k+1)} +  a_{14}^{(k)} \, \varPi_4^{(k+1)} =
  b_1^{(k)} \\
  a_{21}^{(k)}\,\varPi_1^{(k+1)} + a_{22}^{(k)}
  \,\varPi_2^{(k+1)} + a_{25} \, \varPi_5^{(k+1)} =
  b_2^{(k)} \\ 
  a_{33}^{(k)}\,\varPi_3^{(k+1)} + a_{34}^{(k)}
  \,\varPi_4^{(k+1)} = b_3^{(k)} \\ 
  a_{41}^{(k)}\,\varPi_1^{(k+1)} + a_{43}^{(k)}
  \,\varPi_3^{(k+1)} + a_{44}^{(k)}\,\varPi_4^{(k+1)} +
  a_{45}^{(k)}\,\varPi_5^{(k+1)}= b_4^{(k)} \\ 
  a_{52}^{(k)}\,\varPi_2^{(k+1)} + a_{54}\,
  \varPi_4^{(k+1)} + a_{55}^{(k)} \,\varPi_5^{(k+1)}  = b_5^{(k)}
\end{array}
\right\} \qquad \quad \Rightarrow \qquad \quad
\underbrace{
  \left[
    \begin{array}{ccccc}
      a_{11}^{(k)} & a_{12}^{(k)} & 0 & a_{14}^{(k)} & 0 \\ 
      a_{21}^{(k)} & a_{22}^{(k)} & 0 & 0 & a_{25}^{(k)} \\ 
      0   & 0  & a_{33}^{(k)} & a_{34}^{(k)} & 0  \\ 
      a_{41}^{(k)} & 0 & a_{43}^{(k)} & a_{44}^{(k)} &
                                                       a_{45}^{(k)}
      \\  
      0 &  a_{52}^{(k)} & 0 & a_{54}^{(k)} & a_{55}^{(k)}
    \end{array}
  \right]
}_{\mathbb{A}^{(k)}}\,\,\,
\underbrace{
  \left[
    \begin{array}{c}
      \varPi_1^{(k+1)} \\
      \varPi_2^{(k+1)} \\ 
      \varPi_3^{(k+1)} \\ 
      \varPi_4^{(k+1)} \\ 
      \varPi_5^{(k+1)} 
    \end{array}
  \right]
}_{\boldsymbol{\varPi}^{(k+1)}} \,\,\, =\,\,\,
  \underbrace{
    \left[
      \begin{array}{c}
        b_1^{(k)} \\ 
        b_2^{(k)} \\ 
        b_3^{(k)} \\ 
        b_4^{(k)} \\ 
        b_5^{(k)} 
      \end{array}
    \right]
  }_{\boldsymbol{b}^{(k)}}
\end{equation}

Детаљне изразе за коефицијенте $a_{ij}^{(k)}$ и $b_{i}^{(k)}$ можете пронаћи у оквиру презентације. Вредности тих коефијената се мењају током итеративног поступка, и неопходно је решавати овај систем једначина, са новим вредностима коефицијената, све док се не постигне жељена тачност (погледај презентацију). На овом месту ћемо наставити са имплементацијом претходно датог алгоритма у програмски код. 

Прво ћемо позвати модуле који ће нам бити неопходни. То је модул numpy и модул matplotlib који ћемо користити за приказавање одговорајућих дијаграма. Такође, дефинисаћемо и број чворова и број цеви који се налазе у датој мрежи.

```python
import numpy as np
from matplotlib import pyplot as plt
np.set_printoptions(formatter={'float': '{: 0.6f}'.format})
plt.rcParams.update({'font.size': 12})
```

Сада ћемо дефинисати улазне параметре: број чворова, цеви, као и потрошње у чворовима. У програмском коду запремински проток је означен са $Q$ - то је уобичајена ознака у литератури из ове области, уместо $\dot{V}$.

```python
J = 7  # broj cevi
N = 6 # broj cvorova
Qp2 = 0.5
Qp3 = 0.6
Qp5 = 0.4
```

Сходно ознакама које су претходно дефинисане: проток у цевими која повезује чворове $i$ и $j$ ћемо означити са $\dot{V}_{ij} \equiv Q_{ij}$. Сходно овим ознакама, један од начина дефисања протока у нашој мрежи је у виду матрице димензија $N\times N$. Прво ћемо извршити иницијализацију те матрице, тако што ћемо прогласити да су сви њени елементи једнаки нули. Такође, за сваку цев хидрауличку карактеристику (хидраулички отпор) ћемо иницијализовати на исти начин.

```python
Q = np.zeros((N,N))
K = np.zeros((N,N))
```

Сада ћемо навести који од чланова матрице $\mathbb{K}$ су различити од нуле! То су подаци дати у табели.

```python
K[0,1] = 22
K[0,3] = 12
K[1,2] = 11
K[1,4] = 15
K[2,5] = 16
K[3,4] = 9
K[4,5] = 20
```

Затим ћемо извршити иницијализацију почетних вредности протока у деоницама (елементи матрице $\mathbb{Q}$). Индекси елемената матрице који су различити од нуле ће бити исти као претходни индекси елемената матрице $\mathbb{K}$. Треба напоменути да ове вредности протока не морају да задовољавају једначине континуитета!

```python
# Inicijalizacija protoka - u svim deonicama  zapreminski protok je 0.25 m^3/s
Q_0 = 0.75
Q[0,1] = Q_0
Q[0,3] = Q_0
Q[1,2] = Q_0
Q[1,4] = Q_0
Q[2,5] = Q_0
Q[3,4] = Q_0
Q[4,5] = Q_0
```

Сада морамо да дефинишемо особине матрица $\mathbb{K}$ и $\mathbb{Q}$: прва матрица је симетрична, $K_{ji} = K_{ij}$, док је друга матрица антисиметрична, $Q_{ji} = -Q_{ij}$. 

```python
for i in range(N):
    for j in range(N):
        Q[j,i] = - Q[i,j]
        K[j,i] =  K[i,j]
```

Неопходно је дефинисати сада и матрицу-колону која ће одговарати вредностима пијезометарских кота у чворовима. Имамо шест чворова тако да ће матрица-колона имати шест елемената: $\varPi_0, \varPi_1, \dots\varPi_6$. Дефинисаћемо две матрице колоне, јер је неопходно да вршимо поређење вредности пијезометарских кота током итеративног поступка. Сматраћемо да се у оба случаја ради о матрицама колонама чији сви елементи имају исту вредност, једнаку нули.

```python
Pi = np.zeros(N)
Pi_pocetno = np.zeros(N)
```

Коначно, пре почетка прорачуна ћемо дефинисати бројач чија сврха ће бити да сумира број итерација током итеративног поступка, затим да извршимо иницијализацију матрице $\mathbb{A}$, као и да декларишемо листе у које ће смештати вредности пијезометарских кота током итеративног поступка. И да са функцијом print() дефинишемо изглед заглавља при штампању резултата

```python
k = 1
A = np.zeros((5,5))
kota1 = []
kota2 = []
kota3 = []
kota4 = []
kota5 = []
```

Сада ћемо у while() петљи прво одређивати коефицијената матрице $\mathbb{A}$, као и матрице колоне $b$. Потом са командом np.linalg.solve() добити нове вредности пијезометарских кота. На основу тих вредности кота срачунаћемо и нове вредности протока кроз деонице. На крају, морамо да проверимо колико се вредности пијеземетарских кота разликују између суседних итерација: ако је максимална вредност разлике мања од $10^{-6}$ прорачун се прекида, а ако је већи почетне вредности пијезометарских кота се замењују за тренутним, и прорачун се наставља.

```python
Pi_0 = 50.
print("------------------------------------------------------------------------")
print("             Вредности пијезометаских кота током итерација          ")
print("   n\tPi_0\t   Pi_1\t       Pi_2\t  Pi_3\t    Pi_4\t Pi_5")
print("------------------------------------------------------------------------")
while(True):
    ## Елементи матрице А и вектора колоне b
    a12 = - 1/(K[1,2]*np.abs(Q[1,2]))
    a14 = - 1/(K[1,4]*np.abs(Q[1,4]))
    a11 = 1/(K[1,0]*np.abs(Q[1,0])) + np.abs(a12) + np.abs(a14) 
    b1 = Pi_0/(K[1,0]*np.abs(Q[1,0])) - (Q[1,0] + Q[1,2] + Q[1,4])
    
    a21 = a12
    a25 = - 1/(K[2,5]*np.abs(Q[2,5]))
    a22 = np.abs(a21) + np.abs(a25)
    b2 = -2*Qp2 - (Q[2,1] + Q[2,5])
    
    a34 = - 1/(K[3,4]*np.abs(Q[3,4]))
    a33 = 1/(K[3,0]*np.abs(Q[3,0])) + np.abs(a34)
    b3 = Pi_0/(K[3,0]*np.abs(Q[3,0])) - 2*Qp3 - (Q[3,0] + Q[3,4])
    
    a41 = a14
    a43 = a34
    a45 = - 1/(K[4,5]*np.abs(Q[4,5]))
    a44 = np.abs(a41 + a43 + a45)
    b4 = - (Q[4,1] + Q[4,3] + Q[4,5])
    
    a52 = a25
    a54 = a45
    a55 = np.abs(a52 + a54)
    b5 = -2*Qp5 - (Q[5,2] + Q[5,4])
    A = np.array([
        [a11, a12, 0, a14, 0],
        [a21, a22, 0, 0, a25],
        [0, 0, a33, a34, 0],
        [a41, 0, a43, a44, a45],
        [0, a52, 0, a54, a55]
        ])

    b = np.array([b1, b2, b3, b4, b5])          
    x = np.linalg.solve(A, b)
    
    kota1.append(x[0])
    kota2.append(x[1])
    kota3.append(x[2])
    kota4.append(x[3])
    kota5.append(x[4])
    Pi = np.insert(x, 0, Pi_0)
    
   
    print("%4d" %k, str(Pi)[1:-1])
  
    # Nove vrednosti protoka
    Q[0,1] = np.sign(Pi[0] - Pi[1])*np.sqrt(np.abs(Pi[0] - Pi[1])/K[0,1])
    Q[0,3] = np.sign(Pi[0] - Pi[3])*np.sqrt(np.abs(Pi[0] - Pi[3])/K[0,3])
    Q[1,2] = np.sign(Pi[1] - Pi[2])*np.sqrt(np.abs(Pi[1] - Pi[2])/K[1,2])
    Q[1,4] = np.sign(Pi[1] - Pi[4])*np.sqrt(np.abs(Pi[1] - Pi[4])/K[1,4])
    Q[2,5] = np.sign(Pi[2] - Pi[5])*np.sqrt(np.abs(Pi[2] - Pi[5])/K[2,5])
    Q[3,4] = np.sign(Pi[3] - Pi[4])*np.sqrt(np.abs(Pi[3] - Pi[4])/K[3,4])
    Q[4,5] = np.sign(Pi[4] - Pi[5])*np.sqrt(np.abs(Pi[4] - Pi[5])/K[4,5])
    for i in range(N):
        for j in range(N):
            Q[j,i] = - Q[i,j]
    
    
    if(np.max(np.absolute(Pi - Pi_pocetno)) < 1e-6):
        print()
        print("Прорачун је завршен након %d итерација.\n" % k)
        print("----------------------------------------------------------------")
        print("        Вредности запреминских протока Q_{ij}                   ")
        print("----------------------------------------------------------------")
        print (Q)
        break
    else:
        Pi_pocetno = np.copy(Pi)
        k = k + 1
```

Сада ћемо дефинисати резидуале и приказати њихове промене током итеративног поступка, за сваку од пијезометарских кота.

```python
n = np.linspace(1, k-1, k-1)
res1 = []
res2 = []
res3 = []
res4 = []
res5 = []
for i in range(k-1):
    res1.append(np.abs(kota1[i+1] - kota1[i])/kota1[i])
    res2.append(np.abs(kota2[i+1] - kota2[i])/kota2[i])
    res3.append(np.abs(kota3[i+1] - kota3[i])/kota3[i])
    res4.append(np.abs(kota4[i+1] - kota4[i])/kota4[i])
    res5.append(np.abs(kota5[i+1] - kota5[i])/kota5[i])


fig=plt.figure(figsize=(8,6), dpi=100)
ax = fig.add_subplot(111)
ax.set_xlabel("Број итеративних корака", fontsize=14, labelpad=30)
ax.set_ylabel("Резидуали пијезo-кота", fontsize=14, labelpad=30)

plt.plot(n, res1, lw = 1, label='$\Pi_1$')
plt.plot(n, res2, lw = 1, label='$\Pi_2$')
plt.plot(n, res3, lw = 1, label='$\Pi_3$')
plt.plot(n, res4, lw = 1, label='$\Pi_4$')
plt.plot(n, res5, lw = 1, label='$\Pi_5$')
plt.legend()
plt.grid(True, linestyle='dotted')
plt.tight_layout()
```
